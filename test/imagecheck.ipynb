```python
import os
import lightkurve as lk
import matplotlib.pyplot as plt

# Path to your folder with Kepler/TESS FITS files
FOLDER = r"D:\Masrafe\Coding\Git_Hub_code\ml_project\nasa2025\dataset"

# Loop over all FITS files in the folder
for fname in os.listdir(FOLDER):
    if fname.endswith(".fits"):
        fpath = os.path.join(FOLDER, fname)

        try:
            lc = lk.read(fpath).remove_nans().normalize()
            print(f"Plotting: {fname}")

            # Plot
            lc.plot()
            plt.title(fname)
            plt.show()   # show one by one

        except Exception as e:
            print(f"Error reading {fname}: {e}")
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Replace with your file paths
img1 = mpimg.imread("image1.png")   # Real exoplanet transit
img2 = mpimg.imread("image2.png")   # Binary star (false positive)

# Plot side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(img1)
axes[0].set_title("Exoplanet Transit")
axes[0].axis("off")

axes[1].imshow(img2)
axes[1].set_title("Binary Star (False Positive)")
axes[1].axis("off")

plt.tight_layout()
plt.show()
```
